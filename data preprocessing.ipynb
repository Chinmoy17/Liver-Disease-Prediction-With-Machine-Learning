```python
import pandas as pd 
import seaborn as sns

import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("processed_data.csv", encoding='latin1')
```

```python
#label encoding gender column
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the "Gender of the patient" column
df['Gender of the patient'] = label_encoder.fit_transform(df['Gender of the patient'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(subset=['Gender of the patient'], inplace=True)
```

```python
df.info()
```

```python
# Calculate median or mode
# Fill all missing values in the 'Total Bilirubin' column with NaN
df['Total Bilirubin'].fillna(np.nan, inplace=True)
median_total_bilirubin = df['Total Bilirubin'].median()  # or df['Total Bilirubin'].mode()[0] for mode

# Replace missing values with the calculated median or mode
df['Total Bilirubin'].fillna(median_total_bilirubin, inplace=True)
```

```python
df['Direct Bilirubin'].fillna(np.nan, inplace=True)
# Calculate median or mode
median_direct_bilirubin = df['Direct Bilirubin'].median()  # or df['Total Bilirubin'].mode()[0] for mode

# Replace missing values with the calculated median or mode
df['Direct Bilirubin'].fillna(median_direct_bilirubin, inplace=True)
```
